```python
import geopandas as gpd
from shapely.geometry import Point , LineString , Polygon , MultiPoint
import numpy as np
```

```python
Shapefile = gpd.read_file("Layer.shp")
```

```python
i = np.random.randint(0,59)
Numten = Shapefile.geometry[i]
Numten
```

```python
Numten_buf = Numten.buffer(5)
Numten_buf 
```

```python
list1 = []
for k in range(0 , 59):
  P = Numten_buf.intersection(Shapefile.geometry[k])
  list1.append(P)
```

```python
intersect =  Numten_buf.intersection(list1)
list2=[]
for j in range(len(intersect)) :
  if intersect[j].area != 0.0:
    list2.append(intersect[j])
list2
```
