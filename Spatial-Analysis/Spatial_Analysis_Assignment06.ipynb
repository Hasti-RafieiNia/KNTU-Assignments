```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
```

```python
raster_line1 = gdal.Open("line1_PolylineToRaster1.tif")
raster_line3 = gdal.Open("line3_PolylineToRaster1.tif")
```

convert raster to array :

```python
array_line1 = raster_line1.ReadAsArray()
array_line3 = raster_line3.ReadAsArray()
```

shape of array :

```python
array_line1.shape
```

```python
array_line3.shape
```

display of line 1 :

```python
plt.figure(figsize=(5, 5))
plt.imshow(array_line1)
plt.colorbar()
```

display of line 3 :

```python
plt.figure(figsize=(5, 5))
plt.imshow(array_line3)
plt.colorbar()
```

buffer of line 1 :

```python
arr_c1f = array_line1[0:raster_line1.RasterYSize , 0:raster_line1.RasterXSize]
```

```python
list1 = []
for i in range(0 , raster_line1.RasterYSize):
  for j in range(0 , raster_line1.RasterXSize):
    if arr_c1f[i , j] == 0 :
      list1.append((i , j))
```

```python
for k in range(len(list1)):
  arr_c1f[list1[k][0]-2 : list1[k][0]+3 , list1[k][1]-2 : list1[k][1]+3] = 0
```

```python
fig , ax = plt.subplots(figsize = (5, 5))
ax.matshow(arr_c1f)
ax.set_title("buffered line 1")
```

buffer of line 3 :

```python
arr_c3f = array_line3[0:raster_line3.RasterYSize , 0:raster_line3.RasterXSize]
```

```python
list2 = []
for i in range(0 , raster_line3.RasterYSize):
  for j in range(0 , raster_line3.RasterXSize):
    if arr_c3f[i , j] == 0 :
      list2.append((i , j))
```

```python
for k in range(len(list2)):
  arr_c3f[list2[k][0]-2: list2[k][0]+3 , list2[k][1]-2 : list2[k][1]+3] = 0
```

```python
fig , ax = plt.subplots(figsize = (5 , 5))
ax.matshow(arr_c3f)
ax.set_title("buffered line 3")
```

Intersection :

```python
min_x = min(raster_line3.RasterYSize , raster_line1.RasterYSize )
min_y = min(raster_line3.RasterXSize , raster_line1.RasterXSize )
```

```python
intersection = arr_c1f[0:min_x , 0:min_y] + arr_c3f[0:min_x , 0:min_y]
```

```python
fig , ax = plt.subplots(figsize = (10 , 10))
ax.matshow(intersection)
ax.set_title("intersection of 2 lines")
```
