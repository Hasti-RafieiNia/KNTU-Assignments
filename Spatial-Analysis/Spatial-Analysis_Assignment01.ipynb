```python
import math as mt
```

***Part Two : Euclidean Distance***

```python
Euclidean_list = [[[1,2],[3,4]],[[5,6],[7,8]]]
```

```python
def Euclidean_distance(x1, x2, y1, y2) :
  return mt.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
array_Euclidean = []
```

```python
for i in range(len(Euclidean_list)) :
  for j in range(2) :
    for k in range(2) :
      e = Euclidean_distance(Euclidean_list[i][j][k], Euclidean_list[i][j+1][k], Euclidean_list[i][j][k+1], Euclidean_list[i][j+1][k+1])
      array_Euclidean.append(e)
      break
    break
print(array_Euclidean)
```

***Part Two : Manhatan Distance***

0 = blocked points

1 = unblocked points




```python
Manhattan_list = [[[1,2],0] ,[[3,4],1] , [[9,2] , 1] , [[7,8],0] , [[3,7] , 0] ,[[2,3],1] , [[5,2],1]]
```

```python
def manhattan(x , y , b):
    if b == 0:
        return 0 ;
    elif b == 1:
        return([x , y])
```

```python
def manhattan_distance(x , y):
    return abs(x - y)
```

```python
array_manhattan1 = []
```

```python
for i in range(len(Manhattan_list)):
    for j in range(2):
        for k in range(2):
            m1 = manhattan(Manhattan_list[i][j][k] , Manhattan_list[i][j][k+1] , Manhattan_list[i][j+1])
            array_manhattan1.append(m1)
            break
        break        
print(array_manhattan1)
    
```

```python
array_manhattan2 = []
for i in array_manhattan1 :
    if i == 0:
        continue
    array_manhattan2.append(i)
print(array_manhattan2)
```

```python
array_manhattan3 = []
```

```python
for i in range(int(len(array_manhattan2))):
    for j in range(2):
        m2 = manhattan_distance(array_manhattan2[i][j]  , array_manhattan2[i][j+1])
        array_manhattan3.append(m2)
        break

print(array_manhattan3)
```
