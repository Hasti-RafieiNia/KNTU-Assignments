classify the following list of nominal scale.

```python
nominal = np.chararray((8,8))
for i in range(len(nominal)):
    for j in range(len(nominal)):
        nominal[i,j] = chr(np.random.randint(ord('m'), ord('x')))


test_matrix = nominal
res = list(set(i for j in test_matrix for i in j))
res.sort()
print("Unique values in matrix are : " + str(res))



A = np.zeros([8,8] , dtype = int)

for i in range(len(nominal)):
    for j in range(len(nominal)):

        if ( nominal[i , j] == min(res) or nominal[i , j] == res[1] or nominal[i , j] == res[2] ):
            A[i , j] ='1'

        elif (nominal[i , j] == res[3] or nominal[i , j] == res[4] or nominal[i , j] == res[5]):
            A[i , j] ='2'

        elif (nominal[i , j] == res[6] or nominal[i , j] == res[7] or nominal[i , j] == res[8]):
            A[i , j] ='3'

        elif (nominal[i , j] == res[9] or nominal[i , j] == max(res)):
            A[i , j] ='4'

print(A)





arr_LV =A
fig, ax = plt.subplots()
ax.matshow(arr_LV)

for (i, j), z in np.ndenumerate(arr_LV):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```

classify following raster with use of equal interval method, you should implement the method from scratch

```python
arr_c1 = np.random.randint(19, 34, size=(7, 7))

fig, ax = plt.subplots()
ax.matshow(arr_c1)
ax.set_title('Before Classification')
for (i, j), z in np.ndenumerate(arr_c1):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)


test_matrix = arr_c1
res = (list(set(i for j in test_matrix for i in j)))
res.sort()
#print("Unique values in matrix are : " + str(res))

n = 5
d = math.ceil((max(res)-min(res))/n)
B = np.zeros([7,7] , dtype = int)
for i in range(len(arr_c1)):
    for j in range(len(arr_c1)):

        if (min(res) <= arr_c1[i , j] <min(res)+d):
            B[i , j] ='0'

        elif (min(res)+d <= arr_c1[i , j] < min(res)+2*d):
            B[i , j] ='1'

        elif (min(res)+2*d<= arr_c1[i , j] < min(res)+3*d):
            B[i , j] ='2'

        elif (min(res)+3*d <= arr_c1[i , j] < min(res)+4*d):
            B[i , j] ='3'

        elif (min(res)+4*d <= arr_c1[i , j] <= min(res)+5*d):
            B[i , j] ='4'


arr_c1f =B
fig, ax = plt.subplots()
ax.matshow(arr_c1f)
ax.set_title('After Classification')
for (i, j), z in np.ndenumerate(arr_c1f):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```

Wtih use of "Geometrical interval" classify the folloing raster.

```python
arr_c2 = np.random.randint(512, 2156, size=(7, 7))

fig, ax = plt.subplots()
ax.matshow(arr_c2)
ax.set_title('Before Classification')
ax.figure.set_size_inches(7, 7)
for (i, j), z in np.ndenumerate(arr_c2):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
n= 5
test_matrix = arr_c2
res = list(set(i for j in test_matrix for i in j))
res.sort()
#print("Unique values in matrix are : " + str(res))

X = pow((max(res)/min(res)) , 1/n)

C = np.zeros([7,7] , dtype = int)
for i in range(len(arr_c2)):
    for j in range(len(arr_c2)):

        if (min(res) <= arr_c2[i , j] < min(res)*pow(X , 1)) :
            C[i , j] ='0'

        elif (min(res)*pow(X , 1) <= arr_c2[i , j] < min(res)*pow(X , 2)):
            C[i , j] ='1'

        elif (min(res)*pow(X , 2) <= arr_c2[i , j] < min(res)*pow(X , 3)):
            C[i , j] ='2'

        elif (min(res)*pow(X , 3) <= arr_c2[i , j] < min(res)*pow(X , 4)):
            C[i , j] ='3'

        elif (min(res)*pow(X , 4) <= arr_c2[i , j] <= min(res)*pow(X , 5)):
            C[i , j] ='4'


arr_c2f = C
fig, ax = plt.subplots()
ax.matshow(arr_c2f)
ax.set_title('After Classification')
for (i, j), z in np.ndenumerate(arr_c2f):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```
