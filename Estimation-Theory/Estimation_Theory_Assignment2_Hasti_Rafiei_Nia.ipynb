```python
for n in range(1,21):
  V = (np.vander( x1 , n+1 , increasing = True ))
  Xc1 = np.linalg.inv(V.T @ V )@ V.T @ y1
  Yc1 = V @ Xc1
  fig , ax1 = plt.subplots()
  ax1.scatter(x1,y1)
  ax1.plot(x1,Yc1)
```
